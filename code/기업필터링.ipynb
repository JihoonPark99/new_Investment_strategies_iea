## 필터 4 -> 3개에 공통으로 속해진 기업들 

```python
all_li = []
all_li = roe_li.copy()
all_li.extend(roa_li)
all_li.extend(operating_profit_li)
all_li.extend(dept_li)
## 중복 없어
all_li = list(set(all_li))
print(all_li)
```

```python

diff_df = all_df[all_df['종목명'].isin(all_li)]

# diff_df

roa_cond = diff_df['ROA']<1
roe_cond = diff_df['ROE']<1
# operating_cond = all_df['영업이익률'] < 1
dept_cond = diff_df['부채비율'] > 200

diff_df[roa_cond & roe_cond & dept_cond]
```
